```python
url = "https://www.metro.pr/pr/sitemap/sitemap-2019-07.xml"
```

```python
import requests
# https://pythonspot.com/requests-http-for-humans/
```

```python
r = requests.get(url)
print(r.content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.content, 'xml')
soup
```

```python
articles = [ x.string for x in soup.find_all('loc')]
articles
```

```python
len(articles)
```

```python
r = requests.get(articles[1])
soup = BeautifulSoup(r.content, "html.parser")
```

```python
soup.find_all("div", class_="body-content")
```

```python
import re

#print(soup.find_all("div", class_="resumen")[0].text.translate(str.maketrans("aeiou" , ".....")))
txt = soup.find_all("div", class_="resumen")[0].text
txt = re.sub( "[^aeiou ]", "-", txt.lower())
txt = re.sub(  "[aeiou]", ".", txt)
txt
```

```python
print(" ", txt.count(" "))
print(".", txt.count("."))
print("-", txt.count("-"))
print(len(txt))

pattern = [ a + b + c for a in ['-', '.', ' '] for b in ['-', '.', ' '] for c in ['-', '.', ' ']]

[(p, txt.count(p)) for p in pattern]
```

```python
# loop over newspaper articles find the distribution of vowels and consonsants

# distribution of vowels, consonants and spaces looks consistent upon a first (very superficial) check

for n in range(1,51):
    try:
        r = requests.get(articles[n])
        soup = BeautifulSoup(r.content, "html.parser")
        txt = soup.find_all("div", class_="resumen")[0].text
        txt = re.sub( "[^aeiou ]", "-", txt.lower())
        txt = re.sub(  "[aeiou]", ".", txt)

        stats =  [ (" ", txt.count(" ")) , (".", txt.count(".")) , ("-", txt.count("-")) ]

        print(stats)
    except:
        print("Error")
```
